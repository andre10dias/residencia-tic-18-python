```python
print("Caracteres numéricos e seu correspondente código numérico:\n")
for char in '0123456789':
    decimal_value = ord(char)
    print(f"{char} - {decimal_value}")


print("\nCódigo numérico em octal:\n")
for char in '0123456789':
    decimal_value = ord(char)
    octal_value = oct(decimal_value)
    print(f"{char} - {octal_value}")


print("\nCódigo numérico em hexadecimal:\n")
for char in '0123456789':
    decimal_value = ord(char)
    hex_value = hex(decimal_value)
    print(f"{char} - {hex_value}")


print("\nLer um caracter pelo teclado e imprimir seu valor decimal, octal e hexadecimal:\n")
caractere = input("Digite um caractere: ")

if len(caractere) == 1:
    decimal_value = ord(caractere)
    print(f"O valor decimal do caractere '{caractere}' é: {decimal_value}")

    octal_value = oct(decimal_value)
    print(f"O valor octal do caractere '{caractere}' é: {octal_value}")

    hex_value = hex(decimal_value)
    print(f"O valor hexadecimal do caractere '{caractere}' é: {hex_value}")
else:
    print("Por favor, insira apenas um caractere.")


print("\nCaracter 'ç' e 'ã' em decimal, octal e hexadecimal:\n")
caractere_cedilha = 'ç'
caractere_til = 'ã'

decimal_cedilha = ord(caractere_cedilha)
decimal_til = ord(caractere_til)

print(f"O valor decimal do caractere '{caractere_cedilha}' é: {decimal_cedilha}")
print(f"O valor decimal do caractere '{caractere_til}' é: {decimal_til}")

octal_cedilha = oct(decimal_cedilha)
octal_til = oct(decimal_til)

print(f"O valor octal do caractere '{caractere_cedilha}' é: {octal_cedilha}")
print(f"O valor octal do caractere '{caractere_til}' é: {octal_til}")

hex_cedilha = oct(decimal_cedilha)
hex_til = oct(decimal_til)

print(f"O valor hexadecimal do caractere '{caractere_cedilha}' é: {hex_cedilha}")
print(f"O valor hexadecimal do caractere '{caractere_til}' é: {hex_til}")
```
